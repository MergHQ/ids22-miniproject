# Combine all price, generation and load data into one dataframe

```python
import pandas as pd
```

```python
weather = pd.read_csv('./data/weather.csv')
```

```python
import os
```

```python
generation_first = pd.read_csv('./data/Actual Generation per Production Type_201501010000-201601010000.csv')
```

```python
price_first = pd.read_csv('./data/Day-ahead Prices_201501010000-201601010000.csv')
```

```python
load_first = pd.read_csv('./data/Total Load - Day Ahead _ Actual_201501010000-201601010000.csv')
```

```python
generation = pd.DataFrame(columns = generation_first.columns)
```

```python
price = pd.DataFrame(columns = price_first.columns)
```

```python
load = pd.DataFrame(columns = load_first.columns)
```

```python
import glob
```

```python
generation_files = glob.glob('./data/Actual*')
```

```python
generation_files.sort()
```

```python
price_files = glob.glob('./data/Day*')
price_files.sort()
```

```python
load_files = glob.glob('./data/Total*')
load_files.sort()
```

```python
for file in generation_files:
    df = pd.read_csv(file)
    generation = pd.concat([generation, df], axis=0)
    #generation.merge(df, how='outer')
```

```python
generation
```

```python
for file in price_files:
    df = pd.read_csv(file)
    price = pd.concat([price, df], axis=0)
    #generation.merge(df, how='outer')
```

```python
for file in load_files:
    df = pd.read_csv(file)
    load = pd.concat([load, df], axis=0)
    #generation.merge(df, how='outer')
```

```python
price
```

```python
load
```

```python
price.to_csv('./data/all_pricedata.csv')
```

```python
load.to_csv('./data/all_load_data.csv')
```
