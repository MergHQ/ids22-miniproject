```python
df = generation.merge(load, how='outer', left_on = 'MTU', right_on = 'Time (CET/CEST)')
```

Let's replace all n/e values with np.nan

```python
import numpy as np
```

```python
for column in df.columns:
    df.loc[df[column] == 'n/e', column] = np.nan
    df.loc[df[column] == '-', column] = np.nan
```

```python
df
```

Calculate the number of missing values in each column:

```python
for column in df.columns:
    print('number of missing values in column', column)
    print(df[column].isna().sum())
```

There are 8763 values in df for which there is no price data available.

Remove all columns that contain no other data than NaN values:

```python
for column in df.columns:
    if df[column].isna().sum() == len(df.index):
        print('drop column', column)
        df.drop(columns = column, inplace = True)
```

```python
df
```

Create a timestamp from MTU so that I can drop all rows after the date for which we have no data available.

```python
df['start_time'] = df.MTU.str.split('-', expand = True)[0].str.strip()
```

```python
df['end_time'] = df.MTU.str.split('-', expand = True)[1].str.strip()
```

```python
df.head()
```

```python
df['start_time'] = pd.to_datetime(df.start_time, format='%d.%m.%Y %H:%M')
```

```python
df['end_time'] = pd.to_datetime(df.end_time, format='%d.%m.%Y %H:%M')
```

## Change all columns with numerical information to numeric and calculate total generation

```python
gen_columns = list(generation.columns)
```

```python
data = df[df.columns.intersection(gen_columns)]
```

```python
data.drop(columns = ['MTU', 'Area'], inplace = True)
```

```python
for column in data.columns:
    data[column] = pd.to_numeric(data[column])
```

```python
df['generation_total'] = data.sum(axis = 1)
```

```python
for column in data.columns:
    df[column] = pd.to_numeric(df[column])
```
